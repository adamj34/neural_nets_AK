**Expected SARSA**

**Why double expected sarsa works?**

* Temporal Decoupling:
Because the target network is updated less frequently, its parameters represent an older version of the online network. When the online network selects an action (using its current, potentially overestimated Q-values), the evaluation of that action is done using the target network’s parameters, which are “lagging behind” the online network. This lag helps reduce the chance that both networks make the same overestimation error at the same time.

* Error De-correlation:
Overestimation happens when the max operator picks out the highest value among several noisy estimates. If both the selection and evaluation come from the same network, the noise is fully correlated, and the overoptimism compounds. By using a target network that isn’t updated every time the online network is updated, you introduce a level of independence (or at least, a delay) between the noise in the selection process and the noise in the evaluation process. Even if both networks started similarly, the target network’s delayed update means that it doesn’t reflect the most recent (and possibly overestimated) updates, thereby reducing the bias.

```python
# https://arxiv.org/pdf/1704.01665

import numpy as np
import random
import math
from collections import namedtuple, deque

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from plot_helpers import get_graph_mat, plot_solution, plot_rewards

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# The state contains the full distance matrix, node coordinates, and the current tour (partial_solution)
State = namedtuple('State', ('W', 'coords', 'partial_solution'))

def state2tens(state: State) -> torch.Tensor:
    """
    Convert a State into a (N, 5) tensor where for each node we store:
      - Whether it is already in the tour.
      - Whether it is the first node.
      - Whether it is the last visited node.
      - Its x-coordinate.
      - Its y-coordinate.
    """
    solution = set(state.partial_solution)
    sol_last_node = state.partial_solution[-1] if state.partial_solution else -1
    sol_first_node = state.partial_solution[0] if state.partial_solution else -1
    coords = state.coords
    nr_nodes = coords.shape[0]
    xv = [[1 if i in solution else 0,
           1 if i == sol_first_node else 0,
           1 if i == sol_last_node else 0,
           coords[i, 0],
           coords[i, 1]]
          for i in range(nr_nodes)]
    return torch.tensor(xv, dtype=torch.float32, device=device)

Transition = namedtuple('Transition', ('state', 'action', 'reward', 'next_state', 'done'))

class ReplayMemory:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = deque(maxlen=capacity)
    
    def push(self, *args):
        """Save a transition."""
        self.memory.append(Transition(*args))
    
    def sample(self, batch_size) -> list[Transition]:
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)

#############################################
# Define the Q-Network (a simple MLP)
#############################################
class QNetwork(nn.Module):
    def __init__(self, emb_dim=16, T=4):
        """ emb_dim: embedding dimension p
            T: number of iterations for the graph embedding
        """
        super(QNetwork, self).__init__()
        self.emb_dim = emb_dim
        self.T = T
        
        # We use 5 dimensions for representing the nodes' states:
        # * A binary variable indicating whether the node has been visited
        # * A binary variable indicating whether the node is the first of the visited sequence
        # * A binary variable indicating whether the node is the last of the visited sequence
        # * The (x, y) coordinates of the node.
        self.node_dim = 5
        
        # We can have an extra layer after theta_1 (for the sake of example to make the network deeper)
        nr_extra_layers_1 = 2
        
        # Build the learnable affine maps:
        self.theta1 = nn.Linear(self.node_dim, self.emb_dim, True)
        self.theta2 = nn.Linear(self.emb_dim, self.emb_dim, True)
        self.theta3 = nn.Linear(self.emb_dim, self.emb_dim, True)
        self.theta4 = nn.Linear(1, self.emb_dim, True)
        self.theta5 = nn.Linear(2*self.emb_dim, 1, True)
        
        self.theta6 = nn.Linear(self.emb_dim, self.emb_dim, True)
        self.theta7 = nn.Linear(self.emb_dim, self.emb_dim, True)
        
        self.theta1_extras = nn.ModuleList([nn.Sequential(nn.Linear(self.emb_dim, self.emb_dim, bias=True), nn.LayerNorm(self.emb_dim), nn.ReLU())
                                        for _ in range(nr_extra_layers_1)])

    def forward(self, xv, Ws):
        # xv: The node features (batch_size, num_nodes, node_dim)
        # Ws: The graphs (batch_size, num_nodes, num_nodes)
        
        batch_size = xv.shape[0]
        num_nodes = xv.shape[1]

        if Ws.ndim == 2:
            # print(f"Reshaping Ws: {Ws}")
            Ws = Ws.unsqueeze(0)
        Ws = Ws.repeat(batch_size, 1, 1)
        # print(f"Reshaped Ws: {Ws.unsqueeze(3)}")
        
        # pre-compute 1-0 connection matrices masks (batch_size, num_nodes, num_nodes)
        conn_matrices = torch.where(Ws > 0, torch.ones_like(Ws), torch.zeros_like(Ws)).to(device)
        # print(f"Connection matrices: {conn_matrices}")
        
        # Graph embedding
        # Note: we first compute s1 and s3 once, as they are not dependent on mu
        mu = torch.zeros(batch_size, num_nodes, self.emb_dim, device=device)
        s1 = self.theta1(xv)  # (batch_size, num_nodes, emb_dim)
        for layer_suite in self.theta1_extras:
            s1 = layer_suite(s1)  # we apply the extra layer
        
        s3_1 = F.relu(self.theta4(Ws.unsqueeze(3)))  # (batch_size, nr_nodes, nr_nodes, emb_dim) - each "weigth" is a p-dim vector        
        s3_2 = torch.sum(s3_1, dim=1)  # (batch_size, nr_nodes, emb_dim) - the embedding for each node
        s3 = self.theta3(s3_2)  # (batch_size, nr_nodes, emb_dim)
        
        for t in range(self.T):
            s2 = self.theta2(conn_matrices.matmul(mu))    
            mu = F.relu(s1 + s2 + s3)
            
        """prediction"""
        # we repeat the global state (summed over nodes) for each node, 
        # in order to concatenate it to local states later
        global_state = self.theta6(torch.sum(mu, dim=1, keepdim=True).repeat(1, num_nodes, 1))
        
        local_action = self.theta7(mu)  # (batch_dim, nr_nodes, emb_dim)
            
        out = F.relu(torch.cat([global_state, local_action], dim=2))
        return self.theta5(out).squeeze(dim=2).squeeze(dim=0)


class TSP:
    def __init__(self, node_cnt=10):
        self.coords, self.W = get_graph_mat(node_cnt)
        self.memory = ReplayMemory(10000)

    def step(self, state: State, action: int) -> tuple[State, float, bool]:
        """
        Given a state and an action (choosing the next city), compute:
        - The immediate reward (negative distance traveled).
        - The next state.
        - Whether the episode is done (i.e. the tour is complete).
        When the tour is complete, we add the cost to return to the starting city.
        """
        current_city: int = state.partial_solution[-1]
        reward: float = - state.W[current_city, action]
        new_partial: list[int] = state.partial_solution + [action]
        done = False
        if len(new_partial) == state.W.shape[0]:
            # Completed tour; add cost for returning to the starting city.
            reward += - state.W[action, state.partial_solution[0]]
            done = True
        next_state: State = State(state.W, state.coords, new_partial)
        return next_state, reward, done

    def select_action(self, state: State, policy_net, epsilon: float) -> int:
        """
        Selects the next action using an epsilon-greedy policy. Valid actions are the cities not
        already in the partial solution. Returns the index of the selected (max) action.
        """
        valid_actions = list(set(range(state.coords.shape[0])) - set(state.partial_solution))
        if random.random() < epsilon:
            return random.choice(valid_actions)
        else:
            with torch.no_grad():
                state_tensor = state2tens(state) # shape: (nr_nodes, node_dim)
                q_values = policy_net(state_tensor.unsqueeze(0), state.W) # state tensor contains all nodes embeddings at time t
                mask = state_tensor[:, 0].bool() # mask out visited nodes
                q_values_masked = q_values.masked_fill(mask, -float('inf'))
                action = q_values_masked.argmax().item() # return an index of the max value (it's action aka node label)
            return action
        
    def reset(self):
        return State(self.W, self.coords, [0])
    
    def train_tsp_dql(self, policy_net, target_net, num_episodes=500, batch_size=32, gamma=0.9, # note: interesting gamma value
                    epsilon_start=1.0, epsilon_end=0.05, epsilon_decay=200, target_update=10):
        """coords: (n, 2) array of node coordinates, W: (n, n) distance matrix."""
        
        optimizer = optim.AdamW(policy_net.parameters(), lr=1e-3)
        
        episode_rewards = []
        
        for episode in range(num_episodes):
            state = self.reset()
            total_reward = 0.0
            
            # Epsilon decays over episodes.
            # epsilon = epsilon_end + (epsilon_start - epsilon_end) * math.exp(-episode / epsilon_decay)
            epsilon = max(epsilon_end, epsilon_start - (epsilon_start - epsilon_end) * (episode / num_episodes) * 1.1) # linear decay
            done = False
            while not done:
                action = self.select_action(state, policy_net, epsilon)
                next_state, reward, done = self.step(state, action)
                total_reward += reward
                self.memory.push(state, action, reward, next_state, done)
                state = next_state
                
                self.optimize_model(policy_net, target_net, optimizer, batch_size, gamma, epsilon)
                
            episode_rewards.append(total_reward)
            
            # Update the target network periodically.
            if episode % target_update == 0:
                target_net.load_state_dict(policy_net.state_dict())
            
            if episode % 50 == 0:
                print(f"Episode {episode}/{num_episodes}, Total Reward: {total_reward:.2f}, Epsilon: {epsilon:.2f}")
        
        return policy_net, episode_rewards
    
    
    def optimize_model(self, policy_net, target_net, optimizer, batch_size, gamma, epsilon):
        """
        Samples a batch from replay memory and performs one optimization step.
        The target is computed using the target network.
        """
        if len(self.memory) < batch_size:
            return
        
        transitions = self.memory.sample(batch_size)

        batch_s = [t[0] for t in transitions]
        batch_a = [t[1] for t in transitions]
        batch_r = [t[2] for t in transitions]
        batch_next_s = [t[3] for t in transitions]
        batch_done = [t[4] for t in transitions]

        # Convert states to tensors in a single pass
        s_tensors = torch.stack([state2tens(s) for s in batch_s]) # shape: (batch_size, nr_nodes, node_dim)
        q_vals = policy_net(s_tensors, self.W)  # shape: (batch_size, nr_nodes)

        # Gather Q-values for each chosen action
        actions_tensor = torch.tensor(batch_a, dtype=torch.long, device=device)
        chosen_q_vals = q_vals[range(batch_size), actions_tensor]

        # Build target values
        target_vals = []
        with torch.no_grad():
            next_s_tensors = torch.stack([state2tens(ns) for ns in batch_next_s]) # shape: (batch_size, nr_nodes, node_dim)
            q_next_all = target_net(next_s_tensors, self.W)  # shape: (batch_size, nr_nodes)
            for i in range(batch_size):
                if batch_done[i]:
                    target_vals.append(batch_r[i])
                else:
                    # Clone and mask visited nodes
                    q_next_masked = q_next_all[i].clone() # shape: (nr_nodes)
                    mask = torch.zeros_like(q_next_masked, dtype=torch.bool)
                    mask[batch_next_s[i].partial_solution] = True
                    q_next_masked = q_next_masked.masked_fill(mask, -float('inf'))
                    # Expected SARSA
                    available_actions_num = mask.size(0) - mask.sum().item()
                    greedy_action = q_next_masked.argmax().item()
                    greedy_action_val = ((1 - epsilon) + epsilon / available_actions_num) * q_next_masked[greedy_action]
                    non_greedy_actions_val = [(epsilon / available_actions_num) * q_next_masked[non_greedy_action] for non_greedy_action in range(q_next_masked.size(0)) if non_greedy_action != greedy_action and q_next_masked[non_greedy_action] != -float('inf')]
                    expected_sarsa = greedy_action_val + sum(non_greedy_actions_val)
                    
                    target_vals.append(batch_r[i] + gamma * expected_sarsa)

        targets_tensor = torch.tensor(target_vals, dtype=torch.float32, device=device)

        # Compute loss
        loss = F.mse_loss(chosen_q_vals, targets_tensor)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    
    def get_greedy_solution(self, policy_net) -> tuple[list[int], float]:
        """Extracts a tour by always choosing the best action (epsilon=0). Also returns the total length."""
        state = State(self.W, self.coords, [0])
        solution = state.partial_solution.copy()
        tour_length = 0.0
        done = False
        while not done:
            with torch.no_grad():
                state_tensor = state2tens(state)
                q_values = policy_net(state_tensor.unsqueeze(0), self.W)
                # print(f"Q-values: {q_values}")
                q_values_masked = q_values.clone()
                for idx in state.partial_solution:
                    q_values_masked[idx] = -float('inf')
                action = q_values_masked.argmax().item()
            state, reward, done = self.step(state, action)
            tour_length += reward
            solution.append(action)
        return solution, -tour_length
    
    def get_random_solution(self) -> tuple[list[int], float]:
        """Returns a random tour and its total length."""
        random_tour = random.sample(range(self.W.shape[0]), self.W.shape[0])
        tour_length = 0.0
        for i in range(len(random_tour) - 1):
            tour_length += self.W[random_tour[i], random_tour[i+1]]
        # Add the cost to return to the starting city if using a circuit.
        tour_length += self.W[random_tour[-1], random_tour[0]]
        return random_tour, tour_length



if __name__ == '__main__':
    GRAPH_INSTANCES = 20
    EPISODE_NUM=850
    SEED = 4
    NODE_CNT = 15
    random.seed(SEED)
    np.random.seed(SEED)
    torch.manual_seed(SEED)

    # Double learning - two Qnets
    policy_net = QNetwork().to(device)
    target_net = QNetwork().to(device)
    target_net.load_state_dict(policy_net.state_dict())
    target_net.eval()

    rewards_history = []
    for i in range(GRAPH_INSTANCES):
        print(f"################################ Graph Instance: {i} ################################")
        tsp = TSP(node_cnt=NODE_CNT)
        # sync target_net with policy_net for a new graph instance
        target_net.load_state_dict(policy_net.state_dict())

        policy_net, episode_rewards = tsp.train_tsp_dql(policy_net, target_net, num_episodes=EPISODE_NUM)
        rewards_history.append(episode_rewards)

        plot_rewards([-reward.detach().item() for episode_rewards in rewards_history for reward in episode_rewards])
        # plot_rewards([-reward.detach().item() for reward in episode_rewards])

        # random_solution, random_tour_len = tsp.get_random_solution()
        # plot_solution(tsp.coords, random_solution, random_tour_len, "random")

        # greedy_solution, greedy_tour_len = tsp.get_greedy_solution(policy_net)
        # plot_solution(tsp.coords, greedy_solution, greedy_tour_len, "greedy")

    for sample in (range(7)):
        print(f"################################ Sample: {sample} ################################)")
        tsp_instance = TSP(node_cnt=NODE_CNT)
        greedy_solution, greedy_tour_len = tsp_instance.get_greedy_solution(policy_net)
        plot_solution(tsp_instance.coords, greedy_solution, greedy_tour_len, "greedy")

        random_solution, random_tour_len = tsp_instance.get_random_solution()
        plot_solution(tsp_instance.coords, random_solution, random_tour_len, "random")

```
